# Отчет заданий на float

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import subprocess
```

## Ex1

```python
print('N\t Sum')
d = np.zeros(24, dtype='float')
for i in range(1, 25):
    returned_output = subprocess.check_output('ex1.exe '+str(i) + ' 1')
    d[i-1] = float(returned_output.decode('utf-8'))
    print(i, '\t', returned_output.decode('utf-8'))
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(16,6)
fig.suptitle('Ex.1', fontsize=22)

ax.scatter(np.linspace(1,len(d)+1, len(d)), d, marker='D')
ax.plot(np.linspace(1,len(d)+1, len(d)), 1+d-d, '--', color='green')
#ax.set_xlim(0.85, len(d)-1)
ax.grid(True)

ax.set_ylabel('sum', fontsize=18)
ax.set_xlabel('number of steps', fontsize=18)
plt.show()
```

## Ex2

```python
print('Pow. of 2\t Sum')
s =''
d = np.zeros(30, dtype='float')
d_inv = np.zeros(len(d), dtype='float')
for i in range(len(d)):
    n = int(2 ** i)
    returned_output = subprocess.check_output('ex2.exe '+str(n)+' 1')
    returned_output_inv = subprocess.check_output('ex2.exe '+str(n)+' -1')
    d[i] = float(returned_output)
    d_inv[i] = float(returned_output_inv)
    if (len(returned_output.decode('utf-8')) < 6): s = '\t\t'
    else: s = '\t'
    print(i,'\t\t', returned_output.decode('utf-8'), s, returned_output_inv.decode('utf-8'))
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(16,6)
fig.suptitle('Ex.2', fontsize=22)

ax.scatter(np.linspace(1,len(d)+1, len(d)), d, marker='D', label='straight')
ax.scatter(np.linspace(1,len(d_inv)+1, len(d_inv)), d_inv, marker='D', label='inverse')

ax.grid(True)
ax.legend(loc='best', fontsize=16)

ax.set_ylabel('sum', fontsize=18)
ax.set_xlabel('number of steps', fontsize=18)
plt.show()
```

## Ex4

```python
def sin(x, N):
    output = subprocess.check_output('ex4.exe '+str(x)+' '+str(N)+' 1')
    return float(output.decode('utf-8'))
```

```python
#print('N\t Sum')
d = np.zeros(50, dtype='float')
idx = np.linspace(0, 50, len(d))
for i in range(len(idx)):
    d[i] = sin(idx[i], 1000)
    #print(x, '\t', returned_output.decode('utf-8'))
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(16,6)
fig.suptitle('Ex.4, 1000 steps of count', fontsize=22)

ax.scatter(idx, d, marker='D', label='counted')
t = np.linspace(min(idx), max(idx), 1000)
ax.plot(t, np.sin(t), '--', color='green', label='teory')
#ax.set_xlim(0.85, len(d)-1)
ax.grid(True)
ax.set_yscale('log')
ax.set_ylabel('sum, log scale', fontsize=18)
ax.set_xlabel('x', fontsize=18)
ax.legend(loc='best', fontsize=16)
plt.show()
```

## Ex5

```python
def integral(x, n):
    output = subprocess.check_output('ex5.exe '+str(x)+' '+str(n))
    return float(output.decode('utf-8'))
```

```python
d = np.zeros(5, dtype='float')
idx = 2**np.linspace(0, len(d)-1, len(d))
for i in range(len(d)):
    d[i] = integral(8, idx[i])
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(16,6)
fig.suptitle('Ex.5, fixed interval ($x$=8)', fontsize=22)

ax.scatter(idx, d, marker='D')
t = np.linspace(min(idx), max(idx), 10)
ax.plot(t, 0.5*np.sqrt(np.pi)+t-t, '--', color='green')
#ax.set_xlim(0.85, len(d)-1)
ax.grid(True)

ax.set_ylabel('sum', fontsize=18)
ax.set_xlabel('number of steps', fontsize=18)
plt.show()
```

```python
d = np.zeros(5, dtype='float')
idx = 2**np.linspace(0, len(d)-1, len(d))
for i in range(len(d)):
    d[i] = integral(idx[i], 64)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(16,6)
fig.suptitle('Ex.5, fixed amount of steps ($N$=64)', fontsize=22)

ax.scatter(idx, d, marker='D')
t = np.linspace(min(idx), max(idx), 10)
ax.plot(t, 0.5*np.sqrt(np.pi)+t-t, '--', color='green')
#ax.set_xlim(0.85, len(d)-1)
ax.grid(True)

ax.set_ylabel('sum', fontsize=18)
ax.set_xlabel('number of steps', fontsize=18)
plt.show()
```
